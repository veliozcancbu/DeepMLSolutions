```python
"""
Write a Python function that computes the output of the sigmoid activation function given an input value z. The function should return the output rounded to four decimal places.

Example:
Input:
z = 0
Output:
0.5
Reasoning:
The sigmoid function is defined as Ïƒ(z) = 1 / (1 + exp(-z)). For z = 0, exp(-0) = 1, hence the output is 1 / (1 + 1) = 0.5.
"""
```

```python
# NumPy

import math

def sigmoid(z: float) -> float:
    result = 1 / (1 + math.exp(-z))
    return round(result, 4)
```

```python
z = 0
print(sigmoid(z))
```

```python
# PyTorch

import torch

def sigmoid(z: float) -> float:
    z_t = torch.as_tensor(z, dtype=torch.float)
    result = 1 / (1 + torch.exp(-z_t))
    return round(result.item(), 4)
```

```python
z = 0
print(sigmoid(z))
```

```python
# TinyGrad

from tinygrad.tensor import Tensor

def sigmoid_tg(z: float) -> float:
    z_t = z if isinstance(z, Tensor) else Tensor(z)
    result = 1 / (1 + (-z_t).exp())
    return round(float(result.item()), 4)
```

```python
z = 0
print(sigmoid_tg(z))
```
